```python
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
import dataset_preprocess

dataset = dataset_preprocess.get_data_frame()

# shape
print("Shape : ")
print(dataset.shape)

# head
print("\nDataset : ")
print(dataset.head(20))

# descriptions
print("\nDataset Descriptions: ")
print(dataset.describe())

# class distribution
print("\nDataset Classes : ")
print(dataset.groupby('class').size())
```

```python
# box and whisker plots
dataset.plot(kind='box', subplots=True, layout=(404, 1), sharex=False, sharey=False)
plt.show()

# histograms
dataset.hist()
plt.show()

# scatter plot matrix
scatter_matrix(dataset)
plt.show()

print("Piloting completed... ")
```
